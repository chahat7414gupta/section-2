```python
def video_frame_count(my_data):
    # Group the DataFrame by 'video' and count the number of occurrences of each group
    video_counts = my_data.groupby('video').size()
    
    # Convert the resulting Pandas Series to a dictionary
    video_dict = dict(video_counts)
    
    return video_dict
```

```python
my_dict = video_frame_count(my_data)
print(my_dict)
```

```python
def generate_dog_counts(my_data):
    dog_counts = {}
    for i in range(len(my_data)):
        if 'dog' in my_data.loc[i, 'labels']:
            video_name = my_data.loc[i, 'video']
            value = my_data.loc[i, 'value']
            if video_name not in dog_counts:
                dog_counts[video_name] = 0
            dog_counts[video_name] += value
    return dog_counts
```

```python
my_dicte=generate_dog_counts(my_data)
print(my_dicte)
```

```python
def animal_counts(my_data):
    animal_counts = {}
    for index, row in my_data.iterrows():
        if row['video'] == 'video003' or row['value'] % 2 != 0:
            continue
        for animal in row['labels']:
            if animal in animal_counts:
                animal_counts[animal] += 1
            else:
                animal_counts[animal] = 1
    return animal_counts
```

```python
my_dic=animal_counts(my_data)
print(my_dic)
```
